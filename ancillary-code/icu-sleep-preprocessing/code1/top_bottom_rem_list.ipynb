```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
sofa_dir = '/media/mad3/Projects/ICU_SLEEP_STUDY/Sleep_And_Breathing/SOFA_results_studyid'
```

```python
meds = pd.read_csv('/media/mad3/Projects/ICU_SLEEP_STUDY/data/data_analysis/edw_data/edw_icu_sleep_2020_07_07_meds_encounter.csv')
```

```python
top10_rem = pd.read_csv('top10_rem.csv')
bottom10_rem = pd.read_csv('bottom10_rem.csv')
```

```python
def add_meds_and_sofa(top10_rem):

    top10_rem['medications'] = np.nan
    top10_rem['SOFA_mean'] = np.nan
    top10_rem['SOFA_max'] = np.nan

    for jloc, row in top10_rem.iterrows():
        mrn = row.MRN

        try:
            sofa = pd.read_csv(os.path.join(sofa_dir, 'SOFA_ICU' + row.study_id[-3:] + '.csv'))
            top10_rem.loc[jloc, 'SOFA_mean'] = sofa['SOFA_score'].mean()
            top10_rem.loc[jloc, 'SOFA_max'] = sofa['SOFA_score'].max()
        except Exception as e:
            print(e)
            continue
        top10_rem.loc[jloc, 'medications'] = '; '.join(meds[meds.MRN == mrn].MedicationDSC.unique())

    return top10_rem
```

```python
top10_rem = add_meds_and_sofa(top10_rem)
top10_rem.to_csv('top10_rem.csv')
```

```python
bottom10_rem = add_meds_and_sofa(bottom10_rem)
bottom10_rem.to_csv('bottom10_rem.csv')
```
