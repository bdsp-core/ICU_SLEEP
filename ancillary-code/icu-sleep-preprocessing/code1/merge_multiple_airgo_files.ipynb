```python
import pandas as pd
import os
```

```python
single_files = '//mad3/MGH-NEURO-CDAC/Projects/ICU_SLEEP_STUDY/data/data_analysis/AirGo/airgo_research_single_files'
merged_files = '//mad3/MGH-NEURO-CDAC/Projects/ICU_SLEEP_STUDY/data/data_analysis/AirGo/airgo_research'
```

```python
study_id = 6
study_id = str(study_id).zfill(3)
```

```python
files = os.listdir(single_files)
files = [x for x in files if study_id in x]
print(files)
```

```python
data = []
for file in files:
    data.append(pd.read_csv(os.path.join(single_files, file)))
data = pd.concat(data)
data.to_csv(os.path.join(merged_files, file), index=False)
```
