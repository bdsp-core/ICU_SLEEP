```python
savedir_nl = f'/scr/wolfgang/Dropbox (Partners HealthCare)/CircadianRhythms_ICU_Sleep/noise_light'
mad3dir = '/media/mad3/Projects/ICU_SLEEP_STUDY/Data/Enrolled_subjects'

folders = os.listdir(mad3dir)
folders.sort()

for folder in folders:
    
    study_id = folder[:3]
    
    dir_subject = os.path.join(mad3dir, folder)
    folders_subject = os.listdir(dir_subject)
    
    nl_folder = [x for x in folders_subject if 'noise' in x.lower()]
    
    if len(nl_folder) == 0:
        continue
        
    nl_folder = nl_folder[0]
    dir_nl = os.path.join(dir_subject, nl_folder)
    
    print(dir_nl)
    
    savedir_subject = os.path.join(savedir_nl, study_id)
    if not os.path.exists(savedir_subject):
        os.makedirs(savedir_subject)

    for file_tmp in os.listdir(dir_nl):
        try:
            shutil.copyfile(os.path.join(dir_nl, file_tmp), os.path.join(savedir_subject, file_tmp))
        except:
            continue
```
