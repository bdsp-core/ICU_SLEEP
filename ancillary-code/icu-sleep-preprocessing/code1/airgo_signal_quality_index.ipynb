```python
for file in files:
    
    print(file)

    data, hdr, fs = read_in_routine(os.path.join(data_dir, file))
    
    data = data.loc[:, ~data.columns.duplicated()]
    
    data = airgo_signal_quality(data)
    
    # set before/after last valid airgo value to nan:
    valid_band = np.where(np.logical_not(pd.isna(data['band_unscaled'])))[0]
    start = valid_band[0]
    end = valid_band[-1]

    data.loc[:data.iloc[[start]].index[0], 'airgo_signal_quality'] = np.nan
    data.loc[data.iloc[[end]].index[0]:, 'airgo_signal_quality'] = np.nan
    
    write_to_hdf5_file(data, os.path.join(save_dir, file), hdr=hdr)
```
