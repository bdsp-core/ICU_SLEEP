```python
import pandas as pd
import numpy as np
import os
```

```python
summary = pd.DataFrame(columns=['directory', 'annotations_avail', 'duration_avail', 'duration_notna'])
```

```python
natus_dir = '/media/mad3/Datasets_ConvertedData/sleeplab/natus_data'
natus_folders = os.listdir(natus_dir)
folder = natus_folders[1]

for jloc, folder in enumerate(natus_folders):
    
    summary.loc[jloc, 'directory'] = folder
    try:
        ann = pd.read_csv(os.path.join(natus_dir, folder, 'annotations.csv'))
        summary.loc[jloc, 'annotations_avail'] = True
    except:
        summary.loc[jloc, 'annotations_avail'] = False
        continue
    
    if 'duration' in ann.columns:
        summary.loc[jloc, 'duration_avail'] = True
        
        if ann.duration.dropna().shape[0] > 0:
            summary.loc[jloc, 'duration_notna'] = True
        else:
            summary.loc[jloc, 'duration_notna'] = False
        
    else:
        summary.loc[jloc, 'duration_avail'] = False
```

```python
print(f'# Natus Files: {summary.shape[0]}')
print(f'# Annotations available: {summary.annotations_avail.sum()}')
print(f'# duration available: {summary.duration_avail.dropna().sum()}')
print(f'# duration not NAN: {summary.duration_notna.dropna().sum()}')
```

```python
summary.to_csv('natus_annotation_check.csv')
```
