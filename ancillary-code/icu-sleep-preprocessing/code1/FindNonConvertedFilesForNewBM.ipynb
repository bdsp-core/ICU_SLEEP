```python
# LogDirTemp = 'C:/Users/wg984/Wolfgang/ICU-Sleep/'
filep=LogDir+'ConversionStatus_Step1.csv'
ConversionStatus = pd.read_csv(filep)
ConversionStatus['Converted_V4'] = -np.ones([len(ConversionStatus),1],dtype=int)
ConversionStatus.iloc[0,:]
```

```python
# which files have been converted:
```

```python
convertedDir = '//eris1fs2/CDAC/bedmaster_data/batching/ICUSleep_2'
convertedDir2 = '//eris1fs2/CDAC/bedmaster_data/batching/wg984_20191129_163322'
convertedFiles = os.listdir(convertedDir) + os.listdir(convertedDir2)
```

```python
# difference -> which have not been converted
# send those to new reconversion job.
```

```python
cols = ConversionStatus.columns.tolist()
cols = cols[-1:] + cols[:-1]
ConversionStatus= ConversionStatus[cols]
```

```python
ConversionStatus.tail()
```

```python
convertedFileIDs = []
NotconvertedFileIDs = []
for fileID in ConversionStatus.fileID:
    fileIDconverted = np.unique([fileID for x in convertedFiles if fileID in x])
    fileIDNotconverted = np.unique([fileID for x in convertedFiles if not fileID in x])
    if len(fileIDconverted)>0:
        convertedFileIDs.append(fileID)
    elif len(fileIDNotconverted)>0:
        NotconvertedFileIDs.append(fileID)
    
    
    
```

```python
for rowID in range(len(ConversionStatus)):
    if ConversionStatus.iloc[rowID].fileID in convertedFileIDs:
        ConversionStatus['Converted_V4'].iloc[rowID] = 1
    elif ConversionStatus.iloc[rowID].fileID in NotconvertedFileIDs:
        ConversionStatus['Converted_V4'].iloc[rowID] = 0
```
