```python
import numpy as np
import random
random.seed(100)
from sklearn.metrics import roc_auc_score
```

```python
y_true = np.array([0] * 5 + [1] * 5)
y_proba = np.arange(0, 1, 0.1)
y_proba[0] = 0.8 # add some misclassification
y_pred = y_proba > 0.65 # just some threshold to get prediction sample
```

```python
y = np.array([y_true, y_pred, y_proba]).transpose()
```

```python
y
```

### no bootstrapping

```python
accuracy = np.sum(y[:, 0] == y[:, 1]) / y.shape[0]
print(accuracy)
```

```python
roc_auc = roc_auc_score(y[:, 0],  y[:, 2])
print(roc_auc)
```

### with bootstrapping

```python
index_original = np.arange(y.shape[0]).astype(int)
n_bootstraps = 100 # use 10,000 for real dataset
accuracy_all_bootstraps = []
roc_auc_all_bootstraps = []
```

```python
for i_bootstrap in range(n_bootstraps):

    index_bootstrap = random.choices(index_original, k=len(index_original)) # random selection (with replacement) of index
    y_bootstrap = y[index_bootstrap, :]
    accuracy_bootstrap = np.sum(y_bootstrap[:, 0] == y_bootstrap[:, 1]) / y_bootstrap.shape[0]
    roc_auc_bootstrap = roc_auc_score(y_bootstrap[:, 0],  y_bootstrap[:, 2])

    accuracy_all_bootstraps.append(accuracy_bootstrap)
    roc_auc_all_bootstraps.append(roc_auc_bootstrap)
```

```python
roc_auc_all_bootstraps[:10]
```

```python
print(f'Mean [95% CI] Accuracy: {np.mean(accuracy_all_bootstraps)} [{np.quantile(accuracy_all_bootstraps, 0.025)}, {np.quantile(accuracy_all_bootstraps, 0.975)}]')
print(f'Mean [95% CI] ROC AUC: {np.mean(roc_auc_all_bootstraps)} [{np.quantile(roc_auc_all_bootstraps, 0.025)}, {np.quantile(roc_auc_all_bootstraps, 0.975)}]')
```
