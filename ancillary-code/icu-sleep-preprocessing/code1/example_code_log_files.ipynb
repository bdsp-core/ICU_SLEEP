## txt file:

```python
for i in range(2):
    AHI_pred = np.random.rand()
    FP = np.random.rand()
    sensitivity = 0.99999
    all_values_to_save = [AHI_pred, FP, sensitivity]
    with open('samplefile.txt','a') as f:
        for element in all_values_to_save:
            f.write(str(element) + '\t')
        f.write('\n')
```
