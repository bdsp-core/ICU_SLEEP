```python
### Ventilator


df_oxygen = pd.read_csv('/home/wolfgang/repos/Undiagnosed_Apnea/OxygenEDW_all.csv')
df_oxygen['datetime'] = pd.to_datetime(df_oxygen['datetime'])
df_oxygen.set_index('datetime', inplace=True)
```

```python
from datetime import timedelta

summary_subjects_icu['ventilated_pre'] = np.nan
summary_subjects_icu['ventilated_peri'] = np.nan
summary_subjects_icu['ventilated_post'] = np.nan


for jloc, row in summary_subjects_icu.iterrows():
    mrn = row.mrn
    enrollment_date = pd.to_datetime(row.enrolled)
    
    df_mrn = df_oxygen.loc[df_oxygen.MRN == mrn]
    df_mrn = df_mrn.loc[df_mrn['OXYGEN DEVICE'] == 'Ventilator']
    df_mrn_pre = df_mrn.loc[enrollment_date - timedelta(days=31) : enrollment_date]
    df_mrn_peri = df_mrn.loc[enrollment_date : enrollment_date + timedelta(days=14)]
    df_mrn_post = df_mrn.loc[enrollment_date  + timedelta(days=14) : enrollment_date  + timedelta(days=14) + timedelta(days=31)]

    duration_days = 0
    if df_mrn_pre.shape[0] > 0:
        summary_subjects_icu.loc[jloc, 'ventilated_pre'] = 1
        duration_days += (df_mrn_pre.index[-1] - df_mrn_pre.index[0]).total_seconds() / 3600 / 24
    else:
        summary_subjects_icu.loc[jloc, 'ventilated_pre'] = 0

    if df_mrn_peri.shape[0] > 0:
        summary_subjects_icu.loc[jloc, 'ventilated_peri'] = 1
        duration_days += (df_mrn_peri.index[-1] - df_mrn_peri.index[0]).total_seconds() / 3600 / 24
    else:
        summary_subjects_icu.loc[jloc, 'ventilated_peri'] = 0

    if df_mrn_post.shape[0] > 0:
        summary_subjects_icu.loc[jloc, 'ventilated_post'] = 1
        duration_days += (df_mrn_post.index[-1] - df_mrn_post.index[0]).total_seconds() / 3600 / 24
    else:
        summary_subjects_icu.loc[jloc, 'ventilated_post'] = 0
        
    summary_subjects_icu.loc[jloc, 'ventilated_duration_days'] = duration_days

        
summary_subjects_icu['ventilated'] = summary_subjects_icu[['ventilated_pre', 'ventilated_peri', 'ventilated_post']].any(axis=1)
```
