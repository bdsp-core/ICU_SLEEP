```python
def sleep_summary_barplots(ax, data, variables, x_labels,  color='navy', title = ''):
    
    num = len(variables)
    
    means = data[variables].mean(axis=0)
    stds = data[variables].std(axis=0)
    
    if (variables[1] == 'hours_sleep_expert_all'): #manual fix for experts, we don't wanna show hours of sleep twice (total and concordant the same for experts)
        means[1] = 0
        stds[1] = 0

    ax.bar(np.arange(num), means, yerr=[np.zeros(num,), stds], color=color)
    ax.set_xticks(np.arange(num))
    ax.set_xticklabels(x_labels, rotation=90)
    ax.set_title(title)

    for iV in range(num):
        if np.any([x in variables[iV] for x in ['hours']]): 
            decimals = 1
        else: 
            decimals = 0
        if 'hours_sleep' in variables[iV]:
#             offset_text_x = -0.25
            offset_text_x = 0
            offset_text_y = 2
            if iV == 0:
                offset_text_y += 11
        else:
            offset_text_x = 0
            offset_text_y = 2


        var_mean = np.round(data[variables[iV]].mean(), decimals)
        
        # manual correction here for Discordant Sleep = 0. We want to display that as NA in the plots for now.
        if ('discordant' in variables[iV]) & (var_mean == 0):
            var_mean = np.nan

        if (variables[iV] == 'hours_sleep_expert_all') & (iV == 1): #manual fix for experts, we don't wanna show hours of sleep twice (total and concordant the same for experts)
            var_mean = np.nan

        if not np.isnan(var_mean):
            if decimals == 0: 
                    var_mean = int(var_mean)
            var_std = int(np.round(data[variables[iV]].std(),0))
            if (var_mean > 10) & ('hours_sleep' not in variables[iV]):
                ax.annotate(str(var_mean), xy=(iV + offset_text_x, 2), color='white', ha='center', fontweight='normal', fontsize=7, fontfamily='sans-serif')
            else:                
                ax.annotate(str(var_mean), xy=(iV + offset_text_x, var_mean+var_std+offset_text_y), color='black', ha='center', fontweight='normal', fontsize=7, fontfamily='sans-serif')
        else:
            ax.annotate(str('NA'), xy=(iV, 2), fontsize=7, color='black', ha='center')
            
    return ax
```

```python
def set_shared_ylim(ax1, ax2):
    
    ylim_min1, ylim_max1 = ax1.get_ylim()
    ylim_min2, ylim_max2 = ax2.get_ylim()
    ylim_min = np.min([ylim_min1, ylim_min2])
    ylim_min = np.max([ylim_min, 0])
    ylim_max = np.max([ylim_max1, ylim_max2])
    
    ax1.set_ylim(ylim_min, ylim_max)
    ax2.set_ylim(ylim_min, ylim_max)
    
    return ax1, ax2
```

```python

def plot_mean_std_element(data, var, ax, position=0, color='blue', marker=None):
    
    mean_var = data[var].mean()
    std_var = data[var].std()
    ax.errorbar(position, mean_var, yerr=std_var, color=color, fmt='.', zorder=1, marker=marker)
    
    return ax

def plot_median_iqr_element(data, var, ax, position=0, color='blue', marker=None):
    
    mean_var = data[var].median()
    std_var = data[var].quantile(0.75) - data[var].quantile(0.25)
    ax.errorbar(position, mean_var, yerr=std_var, color=color, fmt='.', zorder=1, marker=marker)
    
    return ax

def plot_mean_std_all_elements(var, ax):
    
    ax = plot_mean_std_element(icu_2, var, ax, position=0, color='royalblue')
    ax = plot_mean_std_element(icu_4, var, ax, position=1, color='navy')
    ax = plot_mean_std_element(sleeplab_all, var, ax, position=2, color='gray')
    ax = plot_mean_std_element(sleeplab_ahi_0_5, var, ax, position=3, color='limegreen')
    ax = plot_mean_std_element(sleeplab_ahi_5_15, var, ax, position=4, color='darkgreen')
    ax = plot_mean_std_element(sleeplab_ahi_15_30, var, ax, position=5, color='orange')
    ax = plot_mean_std_element(sleeplab_ahi_30_100, var, ax, position=6, color='red')

    return ax

def plot_median_iqr_all_elements(var, ax):
    
    ax = plot_median_iqr_element(icu_2, var, ax, position=0, color='royalblue')
    ax = plot_median_iqr_element(icu_4, var, ax, position=1, color='navy')
    ax = plot_median_iqr_element(sleeplab_all, var, ax, position=2, color='gray')
    ax = plot_median_iqr_element(sleeplab_ahi_0_5, var, ax, position=3, color='limegreen')
    ax = plot_median_iqr_element(sleeplab_ahi_5_15, var, ax, position=4, color='darkgreen')
    ax = plot_median_iqr_element(sleeplab_ahi_15_30, var, ax, position=5, color='orange')
    ax = plot_median_iqr_element(sleeplab_ahi_30_100, var, ax, position=6, color='red')

    return ax
```

```python
def adjacent_values(vals, q1, q3):
    upper_adjacent_value = q3 + (q3 - q1) * 1.5
    upper_adjacent_value = np.clip(upper_adjacent_value, q3, np.max(vals))

    lower_adjacent_value = q1 - (q3 - q1) * 1.5
    lower_adjacent_value = np.clip(lower_adjacent_value, np.min(vals), q1)
    return lower_adjacent_value, upper_adjacent_value


def set_axis_style(ax, labels):
    ax.get_xaxis().set_tick_params(direction='out')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks(np.arange(1, len(labels) + 1))
    ax.set_xticklabels(labels)
    ax.set_xlim(0.25, len(labels) + 0.75)
    ax.set_xlabel('Sample name')
    
```
