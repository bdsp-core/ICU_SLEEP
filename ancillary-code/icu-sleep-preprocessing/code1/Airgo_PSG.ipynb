#### 1. rename de-identified PSGs to Study ID.

```python
dir_files = '//mad3/MGH-NEURO-CDAC/Projects/AirGo_PSG/Deidentified_edf/'
files_edf = os.listdir(dir_files)
print(len(files_edf))

SleeplabListFile = 'C:/Users/wg984/Dropbox (Partners HealthCare)/AirGoSleepLabData/MasterList_Main_1.23.20.csv'
SleeplabList = pd.read_csv(SleeplabListFile)
```

```python
for studyID in range(358, 500): 

    # 	print(studyID)
    # study = SleeplabList.loc[index]
    study = SleeplabList.where(SleeplabList.SID == 'Air'+str(studyID).zfill(3)).dropna(how='all').squeeze()
    if study.shape[0] == 0: continue
        
    if not isinstance(study.SID, float):
        if study.IncludedInStudy == 0:
            continue

        study_name = study.Path.split('\\')[-1]
        filename = [x for x in files_edf if study_name in x]
        if len(filename) == 0:
            print(f'No file found for {studyID}')
            continue
        filename = filename[0]
        filepath = os.path.join(dir_files, filename)
        filepath_new = os.path.join(dir_files, 'PSG_Air'+str(studyID).zfill(3)+'.edf')
        os.rename(filepath, filepath_new)
```
