```python
ids_to_correct = [112, 114, 115, 116]
time_correction ={112: '2019-07-16 16:17:00',
114: '2019-07-15 18:10:00',
115: '2019-07-15 18:31:00',
116: '2019-07-15 17:25:00'
}
```

```python
airgo_dir = '//mad3/MGH-NEURO-CDAC/Projects/ICU_SLEEP_STUDY/data/data_analysis/AirGo/airgo_research/'
for study_id in [112, 114, 115, 116]:
    airgo_data = pd.read_csv(os.path.join(airgo_dir,'airgo_'+str(study_id)+'.csv'))
    airgo_data.DateTime = pd.to_datetime(airgo_data.DateTime, infer_datetime_format=1)
    print(airgo_data.DateTime.iloc[0])
    new_start_datetime = pd.to_datetime(time_correction[study_id], infer_datetime_format=1)
    print(new_start_datetime)
    timedelta_correction =  new_start_datetime - airgo_data.DateTime.iloc[0]
    print(timedelta_correction)
    airgo_data.DateTime +=  timedelta_correction
    print(airgo_data.head())
    airgo_data.to_csv(os.path.join(airgo_dir,'airgo_'+str(study_id)+'.csv'), index=False)
```
