```python
if not 'PatientID' in cohort.columns:
    
    cohort['PatientID'] = np.nan
    for jloc, row in cohort.iterrows():
        mrn = row.mrn
        cohort.loc[jloc, 'PatientID'] = adt.loc[adt.MRN == mrn, 'PatientID'].unique()[0]
    cohort = cohort[[cohort.columns[-1]] + list(cohort.columns[:-1])]
cohort.rename({'study _id': 'study_id'}, axis=1, inplace=True)
```

```python
cohort_sedation.head(2)
```

```python
cohort['PartOfSedationProject'] = np.nan
for jloc, row in cohort.iterrows():
    patientid = row.PatientID
    cohort.loc[jloc, 'PartOfSedationProject'] = patientid in list(cohort_sedation.PatientID)
```

```python
if 1:
    flowsheets = pd.read_csv(os.path.join(delirium_dir, 'edw_covid_delirium_flowsheets.csv'))
    
    rass_and_cam_keywords = [x for x in flowsheets.FlowsheetMeasureNM.unique() if 'rass' in x.lower()] + [x for x in flowsheets.FlowsheetMeasureNM.unique() if 'cam' in x.lower()]
    rass_and_cam_entries = flowsheets.loc[np.isin(flowsheets.FlowsheetMeasureNM, rass_and_cam_keywords), :]
    rass_and_cam_entries = rass_and_cam_entries[['MRN', 'PatientID', 'PatientEncounterID', 'FlowsheetMeasureNM', 'RecordedDTS', 'MeasureTXT']]

    rass_and_cam_entries['study_id'] = np.nan
    for jloc, row in cohort.iterrows():    
        rass_and_cam_entries.loc[rass_and_cam_entries.MRN == row.mrn, 'study_id'] = row.study_id
    rass_and_cam_entries['study_id'] = rass_and_cam_entries['study_id'].astype(int)
    rass_and_cam_entries = rass_and_cam_entries[[rass_and_cam_entries.columns[-1]] + list(rass_and_cam_entries.columns[:-1])]
    rass_and_cam_entries = rass_and_cam_entries.sort_values(by=['study_id', 'RecordedDTS'])
    rass_and_cam_entries.drop_duplicates(inplace=True)
```
