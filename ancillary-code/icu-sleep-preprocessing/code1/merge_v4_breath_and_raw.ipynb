```python
import numpy as np
import pandas as pd
import os 
```

```python
filedir = 'M:/Projects/ICU_SLEEP_STUDY/data/data_analysis/AirGo/airgo_research'
```

```python
# patients 3,4 and 6 have both v4 breath and raw: merge those:
```

```python
ids = ['003','004','006']
idsel = ids[2]
```

```python
breath = pd.read_csv(os.path.join(filedir, 'airgo_'+idsel+'.csv'))
raw = pd.read_csv(os.path.join(filedir, 'airgo_rawv4_'+idsel+'.csv'))
```

```python
breath = breath[breath.columns[1:]]
breath.head()
```

```python
raw = raw[raw.columns[1:]]
raw.head()
```

```python
merged = pd.concat([breath, raw], axis = 0)
```

```python
merged.head()
```

```python
merged = merged.sort_values('DateTime')
```

```python
merged.head()
```

```python
merged.to_csv(os.path.join(filedir, 'airgo_merged'+idsel+'.csv'), index=False)
```
